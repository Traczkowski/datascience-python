```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
fig, ax = plt.subplots(1, 1)
```

```python
norm.stats(moments='mvsk')
```

```python
norm.ppf(0.01), norm.ppf(0.99)
```

```python
xs = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
ys = norm.pdf(xs)
plt.plot(xs, ys)
```

```python
xs = np.linspace(0, 10, 50)
ys1 = xs ** 2
ys2 = (xs ** 2) + 10
```

```python
plt.plot(xs, ys1, color='#ff33cc', alpha=0.1, label='a')
plt.plot(xs, ys2, color='#33ffcc', alpha=0.9, label='b')
plt.legend()
```

```python
import scipy.stats as scs
```

```python
mu = 60
sd = 5
rv = scs.norm(mu, sd)
xs = np.linspace(mu - (4*sd), mu + (4*sd))
ys = rv.pdf(xs)
plt.plot(xs, ys, color='#ff33cc', alpha=0.75)
plt.axvline(x=65, color='#87cdef')
plt.axhline(y=0.05, color='#cd5711')
```
